

## the last row: start price on 1/2/2024 is USD 16700. We have to find out the predicted end price

## we will ignore the date column. X is the starting price and Y is the ending price

## we will apply simple linear regreesion algorithm Y = mX + C



## we will use supervised ML algorithm. Scikit Learn library has algorithms to solve supervised ML problems (classifications, linear regression etc)

## if you don't have install the libraries: pandas, matplotlib, sklearn

```python
import pandas as pd
from matplotlib import pyplot as plt
import sklearn
```

```python
df = pd.read_csv('nasdaq100.csv')
df
```

```python
# checking for null values in any of the entries
df.isnull().sum()
```

```python
# drop the date column
df2 = df.drop(columns='date')
df2
```

```python
# generate a scatter plot with X and Y data including X & Y labels and a title

plt.scatter(df['starting_price'], df['ending_price'])
plt.xlabel = ('Starting Price')
plt.ylabel('Ending Price')
plt.title('Nasdaq100 Sample Stock Quotes')
```

```python
# DO NOT DROP COLUMN 'ending_price'. this will not make the best fit line go through the red mean dot

x = df2[['starting_price']] 
x.head()
```

```python
y = df2[['ending_price']]
y.head()
```

# Linear Regression
### search for linear regression sklearn

```python
from sklearn.linear_model import LinearRegression # to seel other algorithms press tab after: from sklearn.linear_model import 

reg = LinearRegression()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(), y.mean(), color='red')
plt.scatter(df['starting_price'], df['ending_price'])
plt.xlabel = ('Starting Price')
plt.ylabel('Ending Price')
plt.title('Nasdaq100 Sample Stock Quotes')
```

```python
# train the model

reg.fit(x, y)
```

```python
# value of m of the regression formula Y = mX + C
m = reg.coef_
```

```python
# value of C
c = reg.intercept_
```

```python
# now solve the equation Y =  mX + C for the value of X=16700. The predicted result of y 

y = m * 16700 + c
y
```

```python
# don't need to find m or c or y. use the predict function
reg.predict([[16700]])
```

```python
# check if the best fit line goes through the red dot

df['Predicted_y'] = reg.predict(x)
df
```

```python
plt.plot(x, reg.predict(x))
```

```python
plt.plot(x, reg.predict(x))
plt.scatter(x.mean(), y.mean(), color='red')
plt.scatter(df['starting_price'], df['ending_price'])
plt.xlabel = ('Starting Price')
plt.ylabel('Ending Price')
plt.title('Nasdaq100 Sample Stock Quotes')
```

### Note: the above best fitting line would not go through the red dot when I tried to set the x and y value by dropping columns 
### example: x = df.drop(columns='ending_price')
### simply do: x = df[['ending_price']]

# Loss & Cost



## Loss: every delta changes of x & y is loss.
## Cost: sum of all delta changes of x& y is cost

```python
# check delta loss

df['loss'] = df['ending_price'] - df['Predicted_y']
df
```

## We may use MAE, MSE or RMSE to determine the Cost 

```python
# go to  Scikit learn and copy the libraries for the MAE, MSE and RMSE

from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python
mae = mean_absolute_error(df['ending_price'], df['Predicted_y'])
mae
```

```python
mse = mean_squared_error(df['ending_price'], df['Predicted_y'])
mse
```

```python
rmse = root_mean_squared_error(df['ending_price'], df['Predicted_y'])
rmse
```

```python
# cross check the function valueof mae by calculating manually

df['loss']
```

```python
abs(df['loss'])
```

```python
sum(abs(df['loss']))
```

```python
avg_cost = sum(abs(df['loss']))/len(x)
avg_cost
```

## check performance of the model

```python
performance = reg.score(x, y)
performance
```

## 36% is a poor performance of the model. Look at the scatter plot. values are all over the places
## if you use the reg.predict(x) instead of df['ending_price'] you will see the predicted_y would mostly fall on the best fiteed line

```python
plt.plot(x, reg.predict(x))
plt.scatter(x.mean(), y.mean(), color='red')
plt.scatter(df['starting_price'], df['ending_price'])
plt.scatter(df['starting_price'], reg.predict(x), color='orange')
plt.xlabel = ('Starting Price')
plt.ylabel('Ending Price')
plt.title('Nasdaq100 Sample Stock Quotes')
```

## You can check the performance of a model with other methods. One of them is R**2
## actual - predicted = residual or loss


```python
# will measure performance with r2_score. import r2_score syntax from sklearn
from sklearn.metrics import r2_score
```

```python
# let's revisit the performance with score function. use strating_price or x value and ending_price or y value
performance = reg.score(x, y)
performance
```

```python
# function using ending_price or y and predicted y 
performance2 = r2_score(y, reg.predict(x))
performance2
```
